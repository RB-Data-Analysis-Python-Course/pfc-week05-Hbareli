```python
def factorization(number):
    num=2
    while num<number:
        if (number%num)==0:
            number=number/num
            yield num
            num=2
        else:
            num+=1
    if number!=1:
        yield int(number)

for f in factorization(1386):
    print(f)
```

```python
def fibonacci(max_items):
    a = 1
    b = 1
    numbers = [1, 1]
    while len(numbers) < max_items:
        a, b = b, a + b  # Unpacking
        numbers.append(b)
    return numbers


for number in fibonacci(10):
    print(number)
```
